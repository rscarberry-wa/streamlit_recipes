```python
import ollama
```

```python
models = ollama.list()
for m in models["models"]:
    print(m.model)
```

```python
print(vars(models["models"][0]))
```

```python
model_names = [model.model for model in models["models"]]
print(model_names)
```

```python
def supports_images(model_name: str) -> bool:
    info = ollama.show(model_name)
    return "vision" in info["capabilities"]

def supports_tools(model_name: str) -> bool:
    info = ollama.show(model_name)
    return "tools" in info["capabilities"]

for m in models["models"]:
    model_name = m.model
    if supports_images(model_name):
        print(f"{model_name} supports images")

for m in models["models"]:
    model_name = m.model
    if supports_tools(model_name):
        print(f"{model_name} supports tools")
```
